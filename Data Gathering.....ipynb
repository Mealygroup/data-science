```python
# Defining faker library as fake
fake = Faker()
```

**I'm defining and generating the columns of the dataset.**

```python
# Defining the data
def Generate_mealapp_data(Num_entries):
    Data_entry = []
    
    for _ in range(Num_entries):
        Customer_id = fake.uuid4()
        Dataset = random.choice([Vendor for Vendor in Vendors])
        Vendor = Dataset.get('Name')
        Meal_type = random.choice(list(Dataset.get('Categories').keys()))
        Meal_name = random.choice(list(Dataset.get('Categories')[Meal_type].keys()))
        Meal_id = random.randint(1, 89)
        Order_id = random.randint(1, 100000)
        Meal_price = Dataset.get('Categories')[Meal_type][Meal_name]
        Location = fake.random_element(["Island", "Mainland"])
        Rating = random.randint(1, 5)
        
        if Rating >= 2:
            Review = random.choice(Positive_reviews)
        else:
            Review = random.choice(Negative_reviews)
        
        Meal_description = f"Delicious {Meal_name} made with authentic Nigerian spices and fresh ingredients. It is a popular {Meal_type} dish enjoyed by locals and visitors alike. The {Vendor} prepares this dish with great care and attention to detail, ensuring that it captures the true essence of Nigerian cuisine."
        
        Entry = {
            'Customer_id': Customer_id,
            'Order_id': Order_id,
            'Meal_name': Meal_name,
            'Meal_type': Meal_type,
            'Meal_id': Meal_id,
            'Meal_price': Meal_price,
            'Vendor': Vendor,
            'Rating': Rating,
            'Review': Review,
            'Meal_description': Meal_description,
            'Location': Location
        }
        
        Data_entry.append(Entry)
    
    return Data_entry
```

**I'm generating the rows and saving the dataset into csv file.** 

```python
# Generating 100000 sythentic data 
Mealapp_data = Generate_mealapp_data(100000)

# Save data to CSV file
Filename = 'Mealapp_data.csv'
Fieldnames = ['Customer_id', 'Order_id', 'Meal_id','Meal_name', 'Meal_type', 'Meal_price', 'Vendor', 'Rating', 'Meal_description', 'Review', 'Location']

with open(Filename, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=Fieldnames)
    writer.writeheader()
    writer.writerows(Mealapp_data)

print(f"Mealapp data saved to {Filename}")
```
